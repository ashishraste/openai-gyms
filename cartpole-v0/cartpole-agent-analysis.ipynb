```python
import gym
import numpy as np
from math import log10, radians
from collections import deque
from matplotlib import pyplot as plt
import seaborn as sns

from cartpole_agent import QCartPoleAgent

sns.set()
```

```python
# Plot of adaptive learning rate and exploration probability

min_alpha = 0.1
episode_scaling_factor = 25
def get_alpha(episode_num):
    return max(min_alpha, min(1, 1. - log10((episode_num+1) / episode_scaling_factor)))

alphas = list(map(lambda x : get_alpha(x), [i for i in range(500)]))
sum_i=0
for i in range(len(alphas)):
    if alphas[i] == 0.1: sum_i += 1 
        
plt.plot(alphas)
plt.xlabel('Episode Number')
plt.ylabel('Learning Rate (alpha)')
plt.show()
```
